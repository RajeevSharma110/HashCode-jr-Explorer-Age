```python
from tkinter import *
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
from tkinter import simpledialog
from tkinter import scrolledtext
import time
import fileinput
import sys
```

```python
class BSI:
    flag=0
    line=""
    
    def __init__(self):        
    
        f1 = open("BS.txt","w+")
        L = "Id\t\t"+"Title\t\t"+"Author \t\t"+"year\t\t"+"ISBN\n"
        f1.write(L)
        f1.close()
       
        self.login()
        
    def login(self):
        
        self.login_screen=tk.Tk()
        self.id=tk.IntVar()
        self.title=tk.StringVar()
        self.author=tk.StringVar()
        self.year=tk.IntVar()
        self.isbn=tk.IntVar()
        
       
        self.login_screen.title("Lejhro Book Shop Inventory")
        self.login_screen.geometry("350x350")
        self.username=tk.StringVar()
        self.password=tk.StringVar()
       
        Label(self.login_screen, text="Book Shop Section").pack()
        
        Label(self.login_screen, text="Username").pack()
        self.uname = Entry(self.login_screen, textvariable=self.username)
        self.uname.insert(0,"Tensor")
        self.uname.pack()
        
        Label(self.login_screen, text="").pack()
        Label(self.login_screen, text="Password").pack()
        self.pword = Entry(self.login_screen, textvariable=self.password, show= '*')
        self.pword.insert(0,"4321")
        self.pword.pack()
        
        Label(self.login_screen, text="").pack()
        Button(self.login_screen, text="Login", command= self.valid ,width=10, height=1).pack()
        Button(self.login_screen, text="Quit", command=self.login_screen.destroy,width=10, height=1).pack()  
#         self.login_screen.state('zoomed')
        self.login_screen.mainloop()

    def valid(self):
        u = self.uname.get()
        p = self.pword.get()
        print(u,p)
        if ( u =="Tensor" and p =="4321"):
            messagebox.showinfo("Validation","Login success!!!")
            self.login_screen.withdraw()
            self.book_menu()       
        
        else:
            messagebox.showinfo("Validation","Invalid UserName / Password") 
            self.login_screen.withdraw()
            self.login_screen.iconify()
            self.login_screen.state('normal')       

    def book_menu(self): 
        
        self.Menu_win = tk.Tk()
        self.Menu_win.title("Lejhro Book Shop : Store Section ")
        self.Menu_win.geometry("350x350")
        self.f1 = ttk.Frame(self.Menu_win)
        self.f1.columnconfigure(0, weight=1)
        ttk.Button(self.f1, text='New Book Add',command=self.new_book).grid(column=0, row=1)
        ttk.Button(self.f1, text='View',command=self.book_view).grid(column=0, row=2)
        ttk.Button(self.f1, text='Search', command=self.search).grid(column=0, row=3)
        ttk.Button(self.f1, text='Delete', command=self.delete).grid(column=0, row=4)
        ttk.Button(self.f1, text='Exit',command=self.Menu_win.destroy).grid(column=0, row=5)
        
        for widget in self.f1.winfo_children():
            widget.grid(padx=0, pady=5)
        self.f1.pack()
        self.Menu_win.state('normal')
        self.Menu_win.mainloop()  
   
    def new_book(self):
        
        self.Menu_win.state('withdrawn')      
        self.new_win = tk.Tk()
        self.new_win.title("New Book Detail")  
        self.new_win.geometry("400x450")
        #self.new_win.state('zoomed')
        
        Label(self.new_win, text="Author").pack()
        self.bname = Entry(self.new_win,textvariable=self.id)
        self.bname.focus_set()
        self.bname.pack()
        self.bname.insert(0,"")
        
        
        Label(self.new_win, text="title").pack()
        self.btitle = Entry(self.new_win,textvariable=self.title)
        self.btitle.pack()
        self.btitle.insert(0,"")
        
        Label(self.new_win, text="Year ").pack()
        self.byear = Entry(self.new_win,textvariable=self.year)
        self.byear.pack()
        self.byear.insert(0,"")
        
        Label(self.new_win, text="ISBN ").pack()
        self.bisbn = Entry(self.new_win,textvariable=self.isbn)
        self.bisbn.pack()
        self.bisbn.insert(0,"")
        
    
        Button(self.new_win, text="Save", command= self.upd,width=10, height=1).pack()
        Button(self.new_win, text="Quit", command=self.new_win.destroy,width=10, height=1).pack()    
        Button(self.new_win, text="Back", command=self.stu_menu,width=10, height=1).pack()    
        self.new_win.mainloop()
    
    def modify(self):
         
        self.Menu_win.state('iconic')    
        self.mod_win1 = tk.Tk()
        self.mod_win1.title("Updation of Book Details")  
        self.mod_win1.geometry("400x400")
#         self.mod_win1.state('zoomed')
        
        Button(self.mod_win1, text="Check", command= self.find,width=10, height=1).pack()
                
        self.mod_win1.mainloop()
        
              
    def Search(self):
        
        stn = simpledialog.askstring("Book shop to be Updated!!", "Book name:",parent=self.mod_win1)
        
        self.flag=0
               
        # Read in the file
        filedata = open("BS.txt", "r")               
        for self.line in filedata:
            if stn in self.line:
                self.flag=1
                break 
        filedata.close() 
        
        if self.flag==1:
            
            t=simpledialog.askstring("Input", "title",parent=self.mod_win1)
            a=simpledialog.askstring("Input", "author",parent=self.mod_win1)
            y=simpledialog.askstring("Input", "year",parent=self.mod_win1)
            i=simpledialog.askstring("Input", "isbn",parent=self.mod_win1)
                        
            nl=answer+"\t"+t+"\t"+a+"\t"+y+"\t"+i+"\n"                                
            print(nl)
                        
            
            var1 = l
            var2 = nl
           
            for line in fileinput.input('BS.txt', inplace=1):
                line = line.replace(var1,var2)
                sys.stdout.write(line)
                
            messagebox.showinfo("Message",nl+" Is updated!!!")       
        
        else:
            messagebox.showinfo("Message"," No such student")
            self.mod_win1.withdraw() 
#             self.Menu_win.state('zoomed')
              
        
        
    def upd(self):
       
        
        L = self.bname.get()+self.btitle.get()+self.byear.get()+self.bisbn.get()
        if( L != ""):
            file1 = open("BS.txt","a+")
            L = self.bname.get()+"\t"+self.btitle.get()+"\t"+self.byear.get()+"\t"+self.bisbn.get()+"\n"
            file1.writelines(L)
            file1.close() 
            messagebox.showinfo("Information","Book shop Updated Succesfully!!!") 
            self.new_win.destroy()
            self.new_stu()
        else:
            messagebox.showinfo("Information"," Enter all the data!!! ")
           
            
            
    def getTextInput(self):
        stu_txt=open("BS.txt",'r').read()
        return stu_txt
    
    def book_view(self):
        
        self.Menu_win.state('iconic')
        self.View_win = tk.Tk()
        self.View_win.title("Book Details")  
        self.View_win.geometry("400x450")
        
        self.f2= ttk.Frame(self.View_win)
        
             
        self.stu_text=Text(self.f2,bg='powder blue',width=100,height=50,font=('arial',14,'bold'),padx=5)
        
        self.stu_text.pack()
       
        t=self.getTextInput()
        self.stu_text.insert('insert',t )
            
        
        self.f2.pack()
#         self.View_win.state('zoomed')
        self.Menu_win.state('normal')
        
        self.View_win.mainloop()       
        
```

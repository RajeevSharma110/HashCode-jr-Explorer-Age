```python
from tkinter import *
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
from tkinter import simpledialog
import time
import fileinput
import sys
```

```python
class SRK:
    
    flag=0
    line=""
    
    def __init__(self):        
    
        f1 = open("SR.txt","w+")
        L = "Name\t"+"Gender\t"+"DoB \t"+"Father\t"+"Mother\t"+"ContactNumber\t\t"+"Address\n"
        f1.write(L)
        f1.close()
       
        self.login()
        
    def login(self):
        
        self.login_screen=tk.Tk()
        self.stuname=tk.StringVar()
        self.stugender=tk.StringVar()
        self.studob=tk.StringVar()
        self.stufather=tk.StringVar()
        self.stumother=tk.StringVar()
        self.stucontact=tk.IntVar()
        self.stuaddr=tk.StringVar()
       
        self.login_screen.title("Lejhro School Digital System")
        self.login_screen.geometry("350x350")
        self.username=tk.StringVar()
        self.password=tk.StringVar()
       
        Label(self.login_screen, text="Student Record Section").pack()
        
        Label(self.login_screen, text="Username").pack()
        self.uname = Entry(self.login_screen, textvariable=self.username)
        self.uname.insert(0,"shashi")
        self.uname.pack()
        
        Label(self.login_screen, text="").pack()
        Label(self.login_screen, text="Password").pack()
        self.pword = Entry(self.login_screen, textvariable=self.password, show= '*')
        self.pword.insert(0,"1234")
        self.pword.pack()
        
        Label(self.login_screen, text="").pack()
        Button(self.login_screen, text="Login", command= self.valid ,width=10, height=1).pack()
        Button(self.login_screen, text="Quit", command=self.login_screen.destroy,width=10, height=1).pack()  
        self.login_screen.state('zoomed')
        self.login_screen.mainloop()

    def valid(self):
        u = self.uname.get()
        p = self.pword.get()
        print(u,p)
        if ( u =="shashi" and p =="1234"):
            messagebox.showinfo("Validation","Login success!!!")
            self.login_screen.withdraw()
            self.stu_menu()       
        
        else:
            messagebox.showinfo("Validation","Invalid UserName / Password") 
            self.login_screen.withdraw()
            self.login_screen.iconify()
            self.login_screen.state('normal')       

    def stu_menu(self): 
        
        self.Menu_win = tk.Tk()
        self.Menu_win.title("Lejhro School - Student Record Section ")
        self.Menu_win.geometry("350x350")
        self.f1 = ttk.Frame(self.Menu_win)
        self.f1.columnconfigure(0, weight=1)
        ttk.Button(self.f1, text='New Student Entry',command=self.new_stu).grid(column=0, row=1)
        ttk.Button(self.f1, text='View',command=self.stu_view).grid(column=0, row=2)
        ttk.Button(self.f1, text='Modify', command=self.modify).grid(column=0, row=3)
        ttk.Button(self.f1, text='Exit',command=self.Menu_win.destroy).grid(column=0, row=4)
        
        for widget in self.f1.winfo_children():
            widget.grid(padx=0, pady=5)
        self.f1.pack()
        self.Menu_win.state('normal')
        self.Menu_win.mainloop()  
   
    def new_stu(self):
        
        self.Menu_win.state('withdrawn')      
        self.new_win = tk.Tk()
        self.new_win.title("New Student Detail")  
        self.new_win.geometry("400x450")
        #self.new_win.state('zoomed')
        
        Label(self.new_win, text="Student Name").pack()
        self.sname = Entry(self.new_win,textvariable=self.stuname)
        self.sname.focus_set()
        self.sname.pack()
        self.sname.insert(0,"")
        
        
        Label(self.new_win, text="Gender").pack()
        self.sgender = Entry(self.new_win,textvariable=self.stugender)
        self.sgender.pack()
        self.sgender.insert(0,"")
        
        Label(self.new_win, text="DOB :[ DD/MM/YY ] ").pack()
        self.sdob = Entry(self.new_win,textvariable=self.studob)
        self.sdob.pack()
        self.sdob.insert(0,"")
        
        Label(self.new_win, text="Father Name ").pack()
        self.sfather = Entry(self.new_win,textvariable=self.stufather)
        self.sfather.pack()
        self.sfather.insert(0,"")
        
        Label(self.new_win, text="Mother Name ").pack()
        self.smother = Entry(self.new_win,textvariable=self.stumother)
        self.smother.pack()
        self.smother.insert(0,"")
        
        Label(self.new_win, text="Phone Number").pack()
        self.sphone = Entry(self.new_win,textvariable=self.stucontact)
        self.sphone.pack()
                
        Label(self.new_win, text="Address - Communication:").pack()
        self.saddr = Entry(self.new_win,textvariable=self.stuaddr)
        self.saddr.pack()
        self.saddr.insert(0,"")
               
    
        Button(self.new_win, text="Save", command= self.upd,width=10, height=1).pack()
        Button(self.new_win, text="Quit", command=self.new_win.destroy,width=10, height=1).pack()    
        Button(self.new_win, text="Back", command=self.stu_menu,width=10, height=1).pack()    
        self.new_win.mainloop()
    
    def modify(self):
         
        self.Menu_win.state('iconic')    
        self.mod_win1 = tk.Tk()
        self.mod_win1.title("Updation of Student Details")  
        self.mod_win1.geometry("400x400")
        self.mod_win1.state('zoomed')
        
        Button(self.mod_win1, text="Check", command= self.find,width=10, height=1).pack()
                
        self.mod_win1.mainloop()
        
              
    def find(self):
        
        stn = simpledialog.askstring("Student to be Updated!!", "Student name:",parent=self.mod_win1)
        
        self.flag=0
               
        # Read in the file
        filedata = open("SR.txt", "r")               
        for self.line in filedata:
            if stn in self.line:
                self.flag=1
                break 
        filedata.close() 
        
        if self.flag==1:
            
            g=simpledialog.askstring("Input", "Gender",parent=self.mod_win1)
            dob=simpledialog.askstring("Input", "Date of birth",parent=self.mod_win1)
            fn=simpledialog.askstring("Input", "Father Name",parent=self.mod_win1)
            mn=simpledialog.askstring("Input", "Mother Name",parent=self.mod_win1)
            p=simpledialog.askstring("Input", "Contact",parent=self.mod_win1)
            adr=simpledialog.askstring("Input", "Address",parent=self.mod_win1)
            
            nl=answer+"\t"+g+"\t"+dob+"\t"+fn+"\t"+mn+"\t"+p+"\t"+adr+"\n"                                
            print(nl)
                        
            
            var1 = l
            var2 = nl
           
            for line in fileinput.input('SR.txt', inplace=1):
                line = line.replace(var1,var2)
                sys.stdout.write(line)
                
            messagebox.showinfo("Message",nl+" Is updated!!!")       
        
        else:
            messagebox.showinfo("Message"," No such student")
            self.mod_win1.withdraw() 
            self.Menu_win.state('zoomed')
              
        
        
    def upd(self):
       
        
        L = self.sname.get()+self.sgender.get()+self.sdob.get()+self.sfather.get()+self.smother.get()+self.sphone.get()+self.saddr.get()
        if( L != ""):
            file1 = open("SR.txt","a+")
            L = self.sname.get()+"\t"+self.sgender.get()+"\t"+self.sdob.get()+"\t"+self.sfather.get()+"\t"+ self.smother.get()+"\t"+self.sphone.get()+"\t"+self.saddr.get()+"\n"
            file1.writelines(L)
            file1.close() 
            messagebox.showinfo("Information","Student Record Updated Succesfully!!!") 
            self.new_win.destroy()
            self.new_stu()
        else:
            messagebox.showinfo("Information"," Enter all the data!!! ")
           
            
            
    def getTextInput(self):
        stu_txt=open("SR.txt",'r').read()
        return stu_txt
    
    def stu_view(self):
        
        self.Menu_win.state('iconic')
        self.View_win = tk.Tk()
        self.View_win.title("Student Details")  
        self.View_win.geometry("400x450")
        
        self.f2= ttk.Frame(self.View_win)
        
             
        self.stu_text=Text(self.f2,bg='powder blue',width=100,height=50,font=('arial',14,'bold'),padx=5)
        
        self.stu_text.pack()
       
        t=self.getTextInput()
        self.stu_text.insert('insert',t )
            
        
        self.f2.pack()
        self.View_win.state('zoomed')
        self.Menu_win.state('normal')
        
        self.View_win.mainloop()       
        
 
```
